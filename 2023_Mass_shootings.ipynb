```python
# converting Incident Date column to datetime format
MS_2023["Incident Date"] = pd.to_datetime(MS_2023["Incident Date"], format="%B %d, %Y", errors='coerce')

# droping rows with invalid dates
MS_2023 = MS_2023.dropna(subset=["Incident Date"])

# adding columns for the day of the week and time of day
MS_2023["Day of Week"] = MS_2023["Incident Date"].dt.day_name()
MS_2023["Time of Day"] = pd.cut(MS_2023["Incident Date"].dt.hour, [0, 6, 12, 18, 24], labels=["Night", "Morning", "Afternoon", "Evening"])

# grouping the data by day of the week and time of day and calculate the number of incidents
dataframe_3 = MS_2023.groupby(["Day of Week", "Time of Day"])["Incident ID"].count().unstack()

# creating heatmap
sns.heatmap(dataframe_3, cmap="YlOrBr")
plt.xlabel("Time of Day",fontsize=18, fontweight='bold')
plt.ylabel("Day of Week",fontsize=18, fontweight='bold')
plt.yticks(rotation=90)
plt.title("Number of Gun Violence Incidents by Time of Day and Day of Week in 2023",fontsize=18, fontweight='bold')
plt.show()
```

```python
# creating a dictionary of location types
location_types = {
    "School": ["school", "university"],
    "Workplace": ["workplace"],
    "Public Space": ["street", "park", "mall", "store", "movie theater"],
    "Home": ["home"]
}

# creating a new column to map location types to the data
MS_2023["Location Type"] = MS_2023["Address"].str.lower().apply(lambda x: next((k for k, v in location_types.items() if any(loc in x for loc in v)), "Other"))

# grouping the data by location type and calculate the number of incidents
dataframe_4 = MS_2023.groupby("Location Type")["Incident ID"].count().sort_values()
```

```python
# creating a  bar chart
sns.set(rc={"figure.figsize":(12,12)})
dataframe_4.plot(kind="barh", color="skyblue")
plt.title("Number of Gun Violence Incidents by Location Type in 2023",fontsize=24, fontweight='bold')
plt.xlabel("Number of Incidents",fontsize=18, fontweight='bold')
plt.show()
```
